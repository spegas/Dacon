```python
# Zip the files from given directory that matches the filter
def zipFilesInDir(dirName, zipFileName, filter):
   # create a ZipFile object
   with ZipFile(zipFileName, 'w') as zipObj:
       # Iterate over all the files in directory
       for folderName, subfolders, filenames in os.walk(dirName):
           for filename in filenames:
               if filter(filename):
                   # create complete filepath of file in directory
                   filePath = os.path.join(folderName, filename)
                   # Add file to zip
                   zipObj.write(filePath, basename(filePath))
```

```python
def get_sma_index(pd_stock_price, period_of_list):

    for period in period_of_list:
        sma_column_name = str('SMA_') + str(period)
        pd_stock_price[sma_column_name] = pd_stock_price['Close'].rolling(window=period, min_periods=period_of_list[0]).mean()
        pd_stock_price[sma_column_name] = pd_stock_price[sma_column_name].round(2)

    print(pd_stock_price.head(10))

    return pd_stock_price
```

```python
def get_ema_index(pd_stock_price, period_of_list):

    for period in period_of_list:
        ema_column_name = str('EMA_') + str(period)
        pd_stock_price[ema_column_name] = pd_stock_price['Close'].rolling(window=period, min_periods=period_of_list[0]).mean()
        pd_stock_price[ema_column_name] = pd_stock_price[ema_column_name].round(2)

    print(pd_stock_price.head(10))

    return pd_stock_price
```

```python
def get_mfi_index(pd_stock_price, period_of_list):

    pd_stock_price['TP'] = (pd_stock_price['Close'] + pd_stock_price['High'] + pd_stock_price['Low']) / 3
    pd_stock_price['TP'] = pd_stock_price['TP'].round(2)

    pd_stock_price['PMF'] = 0
    pd_stock_price['NMF'] = 0

    # Loop through the typical price
    for i in range(0, (pd_stock_price.shape[0] - 1)):
        # 당일의 중심가격이 전일의 중심가격보다 크면 긍정적 현금흐름
        if pd_stock_price['TP'].values[i] < pd_stock_price['TP'].values[i + 1]:
            pd_stock_price['PMF'].values[i + 1] = pd_stock_price['TP'].values[i + 1] * pd_stock_price['Volume'].values[i + 1]
            pd_stock_price['NMF'].values[i + 1] = 0
        # 당일의 중심가격이 전일의 중심가격보다 작거나 같으면 부정적 현금흐름
        else:
            pd_stock_price['NMF'].values[i + 1] = pd_stock_price['TP'].values[i + 1] * pd_stock_price['Volume'].values[i + 1]
            pd_stock_price['PMF'].values[i + 1] = 0

    for period in period_of_list:
        pd_stock_price['MFR'] = pd_stock_price['PMF'].rolling(window=period, min_periods=period_of_list[0]).sum() / pd_stock_price['NMF'].rolling(
            window=period, min_periods=period_of_list[0]).sum()
        mfi_columns_name = str('MFI_') + str(period)
        pd_stock_price[mfi_columns_name] = 100 - 100 / (1 + pd_stock_price['MFR'])
        pd_stock_price[mfi_columns_name] = pd_stock_price[mfi_columns_name].round(2)

    # return pd_mfi_index
    return pd_stock_price
```

```python
pd_price = get_mfi_index(pd_price, [5, 10, 20, 30])
```

```python
def get_obv_index(pd_stock_price, period_of_list):

    pd_stock_price['OBV'] = 0

    # Loop through the typical price
    for i in range(1, pd_stock_price.shape[0]):
        # 다음날 가격이 상승한 경우
        if pd_stock_price.Close[i] > pd_stock_price.Close[i - 1]:
            # 현재 obv 값 = 이전 obv + 현재 Volumne 값
            pd_stock_price['OBV'].values[i] = (pd_stock_price['OBV'].values[-1] + pd_stock_price['Volume'].values[i])
        # 다음날 가격이 하락한 경우
        elif pd_stock_price.Close[i] < pd_stock_price.Close[i - 1]:
            # 현재 obv 값 = 이전 obv - 현재 Volumne 값
            pd_stock_price['OBV'].values[i] = (pd_stock_price['OBV'].values[-1] - pd_stock_price['Volume'].values[i])
        else:
            pd_stock_price['OBV'].values[i] = pd_stock_price['OBV'].values[-1]

    for period in period_of_list:
        obv_columns_name = str('OBV_') + str(period)
        pd_stock_price[obv_columns_name] = pd_stock_price['OBV'].ewm(com=period, min_periods=period_of_list[0]).mean()
        pd_stock_price[obv_columns_name] = pd_stock_price[obv_columns_name].round(2)

    return pd_stock_price
```

```python
pd_price = get_obv_index(pd_price, [5, 10, 20, 30])
```

```python
def get_pvt_index(pd_stock_price, period_of_list):

    pd_stock_price['PVT'] = 0

    # Loop through the typical price
    for i in range(1, pd_stock_price.shape[0]):
        pvt_value = pd_stock_price['Close'].values[i] - pd_stock_price['Close'].values[i - 1]
        pvt_value = (pvt_value / pd_stock_price['Close'].values[i - 1]) * pd_stock_price['Volume'].values[i]
        pd_stock_price['PVT'].values[i] = pvt_value + pd_stock_price['PVT'].values[-1]

    for period in period_of_list:
        pvt_columns_name = str('PVT_') + str(period)
        pd_stock_price[pvt_columns_name] = pd_stock_price['PVT'].ewm(com=period, min_periods=period_of_list[0]).mean()
        pd_stock_price[pvt_columns_name] = pd_stock_price[pvt_columns_name].round(2)

    return pd_stock_price
```

```python
def get_macd_index(pd_stock_price, m_NumFast=12, m_NumSlow=26, m_NumSignal=9):

    pd_stock_price['EMAFast'] = pd_stock_price['Close'].ewm(span=m_NumFast, min_periods=m_NumFast - 1).mean()
    pd_stock_price['EMASlow'] = pd_stock_price['Close'].ewm(span=m_NumSlow, min_periods=m_NumSlow - 1).mean()
    pd_stock_price['MACD'] = pd_stock_price['EMAFast'] - pd_stock_price['EMASlow']

    return pd_stock_price
```

```python
def get_bollinger_bands(pd_stock_price, period_of_list):

    for period in period_of_list:
        bl_ma_columns_name = str('BL_MA_') + str(period)
        pd_stock_price[bl_ma_columns_name] = pd_stock_price['Close'].ewm(com=period, min_periods=period_of_list[0]).mean()
        pd_stock_price[bl_ma_columns_name] = pd_stock_price[bl_ma_columns_name].round(2)

        bl_std_columns_name = str('BL_STD_') + str(period)
        pd_stock_price[bl_std_columns_name] = pd_stock_price['Close'].ewm(com=period, min_periods=period_of_list[0]).std()
        pd_stock_price[bl_std_columns_name] = pd_stock_price[bl_std_columns_name].round(2)

        bl_up_columns_name = str('BL_UP_') + str(period)
        pd_stock_price[bl_up_columns_name] = pd_stock_price[bl_ma_columns_name] + (2 * pd_stock_price[bl_std_columns_name])

        bl_lower_columns_name = str('BL_LO_') + str(period)
        pd_stock_price[bl_lower_columns_name] = pd_stock_price[bl_ma_columns_name] - (
                    2 * pd_stock_price[bl_std_columns_name])

    return pd_stock_price
```

```python
pd_price = get_bollinger_bands(pd_price, [5, 10, 20, 30])
```

```python
def get_stochastic_oscillator(pd_stock_price, period_of_list):

    for period in period_of_list:
        st_max_columns_name = str('ST_MAX_') + str(period)
        pd_stock_price[st_max_columns_name] = pd_stock_price['High'].rolling(period, min_periods=period_of_list[0]).max()

        st_min_columns_name = str('ST_MIN_') + str(period)
        pd_stock_price[st_min_columns_name] = pd_stock_price['Low'].rolling(period, min_periods=period_of_list[0]).min()

        st_k_columns_name = str('ST_K_') + str(period)
        pd_stock_price[st_k_columns_name] = (pd_stock_price['Close'] - pd_stock_price[st_min_columns_name]) * 100
        pd_stock_price[st_k_columns_name] = pd_stock_price[st_k_columns_name] / (
                    pd_stock_price[st_max_columns_name] - pd_stock_price[st_min_columns_name])

        st_d_columns_name = str('ST_D_') + str(period)
        pd_stock_price[st_d_columns_name] = pd_stock_price[st_k_columns_name].rolling(3).mean()

    return pd_stock_price
```

```python
pd_price = get_stochastic_oscillator(pd_price, [5, 10, 20, 30])
```

```python
def get_rate_of_price_change(pd_stock_price, period_of_list):

    for period in period_of_list:
        price_diff_columns_name = str('R_Price_diff_') + str(period)
        pd_stock_price[price_diff_columns_name] = pd_stock_price['Close'].diff(period - 1)

        price_shift_columns_name = str('R_Price_shift_') + str(period)
        pd_stock_price[price_shift_columns_name] = pd_stock_price['Close'].shift(period - 1)

        price_roc_columns_name = str('R_Price_ROPC_') + str(period)
        pd_stock_price[price_roc_columns_name] = pd_stock_price[price_diff_columns_name] / pd_stock_price[price_shift_columns_name]

    return pd_stock_price
```

```python
pd_price = get_rate_of_price_change(pd_price, [5, 10, 20, 30])
```

```python
def get_rate_of_volume_change(pd_stock_price, period_of_list):

    for period in period_of_list:
        price_diff_columns_name = str('R_Volume_diff_') + str(period)
        pd_stock_price[price_diff_columns_name] = pd_stock_price['Volume'].diff(period - 1)

        price_shift_columns_name = str('R_Volume_shift_') + str(period)
        pd_stock_price[price_shift_columns_name] = pd_stock_price['Volume'].shift(period - 1)

        price_roc_columns_name = str('R_Volume_ROPC_') + str(period)
        pd_stock_price[price_roc_columns_name] = pd_stock_price[price_diff_columns_name] / pd_stock_price[price_shift_columns_name]

    return pd_stock_price
```

```python
def get_cci_index(pd_stock_price, period_of_list):

    for period in period_of_list:
        period_sma_columns_name = str('CCI_TP_SMA_') + str(period)
        pd_stock_price[period_sma_columns_name] = pd_stock_price['TP'].rolling(period, min_periods=period_of_list[0]).mean()

        period_mad_columns_name = str('CCI_TP_MAD_') + str(period)
        pd_stock_price[period_mad_columns_name] = pd_stock_price['TP'].rolling(period, min_periods=period_of_list[0]).apply(lambda x: pd.Series(x).mad())

        period_cci_columns_name = str('CCI_') + str(period)
        pd_stock_price[period_cci_columns_name] = (pd_stock_price['TP'] - pd_stock_price[period_sma_columns_name])
        pd_stock_price[period_cci_columns_name] = pd_stock_price[period_cci_columns_name] / (0.015 * pd_stock_price[period_mad_columns_name])

    return pd_stock_price
```

```python
def get_adx_index(pd_stock_price, period_of_list):

    pd_stock_price['HL'] = pd_stock_price['High'] - pd_stock_price['Low']
    pd_stock_price['absHC'] = abs(pd_stock_price['High'] - pd_stock_price['Close'].shift(1))
    pd_stock_price['absLC'] = abs(pd_stock_price['Low'] - pd_stock_price['Close'].shift(1))
    pd_stock_price['TR'] = pd_stock_price[['HL', 'absHC', 'absLC']].max(axis=1)

    pd_stock_price['UpMove'] = pd_stock_price['High'] - pd_stock_price['High'].shift(1)
    pd_stock_price['DonwMove'] = pd_stock_price['Low'].shift(1) - pd_stock_price['Low']
    pd_stock_price['Zero'] = 0

    pd_stock_price['PlusDM'] = np.where((pd_stock_price['UpMove'] > pd_stock_price['DonwMove']) &
                                          (pd_stock_price['UpMove'] > pd_stock_price['Zero']),
                                          pd_stock_price['UpMove'], 0)
    pd_stock_price['MinusDM'] = np.where((pd_stock_price['UpMove'] < pd_stock_price['DonwMove']) &
                                           (pd_stock_price['DonwMove'] > pd_stock_price['Zero']),
                                           pd_stock_price['DonwMove'], 0)

    for period in period_of_list:

        atr_column_name = str('ATR_') + str(period)
        pd_stock_price[atr_column_name] = pd_stock_price['TR'].rolling(window=period, min_periods=period_of_list[0]).mean()

        natr_column_name = str('NATR_') + str(period)
        pd_stock_price[natr_column_name] = (pd_stock_price[atr_column_name] / pd_stock_price['Close']) * 100

        plus_di_column_name = str('plusDI_') + str(period)
        pd_plus_di_values = (pd_stock_price['PlusDM'] / pd_stock_price[atr_column_name])
        pd_stock_price[plus_di_column_name] = 100 * pd_plus_di_values.ewm(span=period, min_periods=0, adjust=True, ignore_na=False).mean()

        minus_di_column_name = str('MinusDI_') + str(period)
        pd_minus_di_values = (pd_stock_price['MinusDM'] / pd_stock_price[atr_column_name])
        pd_stock_price[minus_di_column_name] = 100 * pd_minus_di_values.ewm(span=period, min_periods=0, adjust=True, ignore_na=False).mean()

        adx_column_name = str('ADX_') + str(period)
        sub_plus_di_minus_di = (pd_stock_price[plus_di_column_name] - pd_stock_price[minus_di_column_name])
        add_plus_di_minus_di = (pd_stock_price[plus_di_column_name] + pd_stock_price[minus_di_column_name])

        pd_stock_price[adx_column_name] = 100 * (abs(sub_plus_di_minus_di / add_plus_di_minus_di)).ewm(span=period, min_periods=0, adjust=True, ignore_na=False).mean()

    return pd_stock_price
```
